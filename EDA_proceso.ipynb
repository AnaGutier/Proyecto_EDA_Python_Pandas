## Primera inspección de los datos y organización de los mismos

Como cuento con dos archivos diferentes, procedo a combiar los datos para tener toda la información en un mismo lugar.

El archivo excel tiene varias hojas que corresponden a los años 2012, 2013 y 2014, contando con las mismas columnas en todas ellas. 
Para un mejor tratamiento de los datos decido juntarlas todas en un mismo dataframe.

Tras esto, cambino los datos de los dos archivos mediante las columnas de identificación (ID)
Para ello primero estandarizo las columnas ID y id_ y los fusiono, comprobando lugo el resultado.

```python

# Uno las hojas del archivo excel
excel_file = 'datos_proyecto/customer-details.xlsx'
# Leer todas las hojas (para que resulte un diccionario de DataFrames)
dfs = pd.read_excel (excel_file, sheet_name = None)
# Unir todas las hojas en un único DataFrame
df_fullcustomer = pd.concat(dfs.values(), ignore_index = True)

# Muestro el resultado
df_fullcustomer
```

```python
# Estandarizo las columnas de identificación como id_cliente

df_fullcustomer.rename(columns={'ID': 'id_cliente'}, inplace=True)
df_bank.rename(columns={'id_': 'id_cliente'}, inplace=True)
```

```python
# Fusiono ambos en un único dataframe
# Fusionar los dos dataframes usando la columna 'id_cliente'
df_completo = pd.merge(df_bank, df_fullcustomer, on='id_cliente', how='inner')

# Muestro las primeras 5 filas para ver cómo quedó la fusión
print("Primeras filas del DataFrame fusionado:")
pd.set_option('display.max_columns', None) # Para que se muestren todas las columnas
display(df_completo.head())
```

```python
# Realizo una inspección general del dataframe para una mejor comprensión inicial de los datos

# Muestro las dimensiones del DataFrame (filas, columnas)
print(f"\nDimensiones del DataFrame: {df_completo.shape}")

# Muestro estadísticas descriptivas para las columnas numéricas
print("\nEstadísticas descriptivas:")
display(df_completo.describe())
```

```python

# Obtengo un resumen conciso, incluyendo tipos de datos y valores no nulos
print("\nInformación general del DataFrame:")
df_completo.info()
```

## Limpieza y transformación de datos

Con la inspección general de los datos ya se aventuran algunas correcciones necesarias a realizar.

En primer lugar observo que hay columnas redundates como 0_x o 0_y que no aportan información relevante. 

Además, hay datos con un tipo de dato que no les corresponde, por ejemplo las de tipo objet que deberían de ser numéricas.
Para solucionarlo reemplazo las comas por un punto para convertir los datos a float.
A su vez, la columna date (que está en español) necesita tratarse para poder convertirse a datatime. 

Otro elemento importante son las columnas con un número de datos menor a 43000, que tienen valores faltantes. Además, es posible que surgen valores Nan a corregir.
- Hay alfunas columnas con pocos ausentes: job, marital, education, housing, loan y date, que tienen relativamente pocos valores nulos, no supone un problema eliminar las filas que los contienen para no introducir ruido en el análisis.
- Para las columnas numéricas (age, euribor3m, etc.) puedo rellenar los NaN con la media o la mediana. La mediana es generalmente más robusta ante valores atípicos.
- Columna default: Esta columna tiene muchos valores ausentes. Antes de decidir, voy a ver qué valores contiene.


```python
# Elimino las dos columnas que no aportan información

df_completo.drop('Unnamed: 0_x',axis = 1,inplace = True)

df_completo.drop('Unnamed: 0_y',axis = 1,inplace = True)
```

```python
# Cambio los nomnres de las columnas que tienen puntos en su nombre para evitar errores
df_completo.columns = df_completo.columns.str.replace('.', '_', regex=False)
print("Nombres de columna corregidos:")
print(df_completo.columns)
```

```python
# Columnas a convertir a numérico (con los nombres ya corregidos)
cols_to_numeric = ['cons_price_idx', 'cons_conf_idx', 'euribor3m', 'nr_employed']

for col in cols_to_numeric:
    # Convierto a string, remplazo la coma, y convierto a float
    # errors='coerce' convertirá los valores que no se puedan transformar en NaN (Not a Number)
    df_completo[col] = df_completo[col].astype(str).str.replace(',', '.', regex=False).astype(float)

# Conversión de la columna 'date'
# Mapeo los meses de español a inglés
meses = {
    'enero': 'January', 'febrero': 'February', 'marzo': 'March', 'abril': 'April',
    'mayo': 'May', 'junio': 'June', 'julio': 'July', 'agosto': 'August',
    'septiembre': 'September', 'octubre': 'October', 'noviembre': 'November', 'diciembre': 'December'
}

# Me aseguro de que la columna 'date' sea de tipo string para poder usar .str
df_completo['date'] = df_completo['date'].astype(str)

# Cambio los nombres de los meses
df_completo['date_ingles'] = df_completo['date'].str.split('-').str[1].map(meses)
df_completo['date_formateada'] = df_completo['date'].str.split('-').str[0] + '-' + df_completo['date_ingles'] + '-' + df_completo['date'].str.split('-').str[2]

# Convierto a datetime
# 'errors=coerce' es útil aquí por si alguna fecha está mal formateada y no se puede convertir
df_completo['date'] = pd.to_datetime(df_completo['date_formateada'], format='%d-%B-%Y', errors='coerce')

# Elimino las columnas auxiliares que ya no necesito
df_completo = df_completo.drop(columns=['date_ingles', 'date_formateada'])

# Verifico los cambios
print("Tipos de datos después de la corrección:")
df_completo.info()
```

```python
# 1. Columnas con pocos nulos: elimino las filas con nulos en estas columnas.
cols_to_drop_na = ['job', 'marital', 'education', 'housing', 'loan', 'date']
df_completo.dropna(subset=cols_to_drop_na, inplace=True)

# 2. Columnas numéricas: añado con la mediana.
for col in ['age', 'euribor3m', 'cons_price_idx', 'cons_conf_idx', 'nr_employed']:
    median_val = df_completo[col].median()
    df_completo[col].fillna(median_val, inplace=True)

# 3. Columna 'default': compruebo cuál es la situación primero.
print("Valores en la columna 'default':")
print(df_completo['default'].value_counts(dropna=False))

# La gran mayoría son 0.0 o NaN. La categoría '1.0' es muy rara.
# Una opción es asumir que los NaN son 'no' (0.0) o crear una categoría 'unknown'.
# Por simplicidad y teniendo en cuenta que casi no hay 'default=1', relleno con 0.
df_completo['default'].fillna(0.0, inplace=True)


# Verificación final de nulos
print("\nResumen de valores nulos tras la limpieza:")
print(df_completo.isnull().sum())
```

### Creación de nuevas columnas
El número de hijos totales puede ser un dato interesante, así que creo una columna sumano kidhome y teenhome.
Además, para tener un mejor manejo de los grupos sociales según edad decido agrupar estas en categorías: joven - adulto - adulto mayor - senior.

```python
# 1. Creo la columna 'total_children'
df_completo ['total_children'] = df_completo['Kidhome'] + df_completo['Teenhome']

# 2. Creo los grupos de edad ('age_group')
# Defino los límites y las etiquetas para los grupos de edad
bins = [17, 30, 45, 60, 100]
labels = ['Joven (17-30)', 'Adulto (31-45)', 'Adulto Mayor (46-60)', 'Senior (61+)']
df_completo ['age_group'] = pd.cut(df_completo['age'], bins=bins, labels=labels, right=False)

# Muestro las nuevas columnas junto con las originales para verificar
print("Verificación de las nuevas columnas creadas:")
display(df_completo[['age', 'age_group', 'Kidhome', 'Teenhome', 'total_children']].head())
```

## Exploración de datos

Para comenzar con el análisis de los datos, haré una fase de variables categóricas y otra con las numéricas, para entender cada variable en sí misma y posteriormente relacioanrlas. Las variables categóricas me ayudarán a entender el perfil de los clientes (su estado civil, su trbajo, nivel educativo, o si aceptaron o no la oferta y).

La columna y es de vital importancia porque inica si el cliente compró o no el producto, y muestra el tipo de conjunto de datos con el que trato. Para comprenderlo mejor, importo librerías de visualización (matplot y seaborn).

```python
# Importo librerías de visualización
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Configuración de estilo para los gráficos
sns.set_style('whitegrid')

# Calculo el conteo y el porcentaje de cada categoría en 'y'
y_counts = df_completo['y'].value_counts()
y_percentages = df_completo['y'].value_counts(normalize=True) * 100

print("Distribución de la variable objetivo 'y':")
print(y_counts)
print("\nPorcentaje de la variable objetivo 'y':")
print(y_percentages)

# Visualización
plt.figure(figsize=(8, 6))
sns.countplot(x='y', data=df_completo, palette='viridis')
plt.title('Distribución de Suscripciones (Variable "y")', fontsize=16)
plt.xlabel('¿Cliente ha suscrito?', fontsize=12)
plt.ylabel('Número de Clientes', fontsize=12)
plt.show()
```

Hay una diferencia muy grande entre el sí y el no, predominando este último por lo que paso a estudiar el resto de varibales, en primer lugar las categóricas y las numéricas por separado para después relacionarlas. Incluyo elementos como boxplots para atender a los posibles valores atópicos a tener en cuenta.

```python
# Análisis de otras variables categóricas

# Lista de columnas categóricas a analizar
categorical_cols = ['job', 'marital', 'education', 'age_group']

# Creo un gráfico para cada columna
for col in categorical_cols:
    plt.figure(figsize=(12, 7))
    sns.countplot(y=col, data=df_completo, order=df_completo[col].value_counts().index, palette='plasma')
    plt.title(f'Distribución de Clientes por "{col}"', fontsize=16)
    plt.xlabel('Número de Clientes', fontsize=12)
    plt.ylabel(col.capitalize(), fontsize=12)
    plt.show()
```

```python
# Análisis de variables numéricas

# Lista de columnas numéricas a analizar
numerical_cols = ['age', 'Income', 'duration']

# Creo un histograma y un boxplot para cada una
for col in numerical_cols:
    plt.figure(figsize=(14, 5))

    # Histograma
    plt.subplot(1, 2, 1)
    sns.histplot(df_completo[col], kde=True, bins=30, color='skyblue')
    plt.title(f'Distribución de {col}', fontsize=14)
    
    # Boxplot
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df_completo[col], color='lightgreen')
    plt.title(f'Boxplot de {col}', fontsize=14)
    
    plt.suptitle(f'Análisis de la variable "{col}"', fontsize=18)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
```

Una vez consideradas las varibales de forma individual, realizo un análisis bivariado en realiacón a la variable principal que en este caso es y (la suscripción o no del producto). Esto me permite comparar la proporción del sí y el no en cada categoría, para idenficar el tipo de cliente más propicio al sí. 

```python
# Análisis Bivariado: Relación con la Suscripción ('y')

# Analizo variables categóricas vs 'y'
bivariate_cols = ['job', 'marital', 'education', 'age_group', 'housing', 'loan']

for col in bivariate_cols:
    # Usamos 'hue' para separar por la variable 'y'
    plt.figure(figsize=(12, 7))
    sns.countplot(y=col, data=df_completo, order=df_completo[col].value_counts().index, hue='y', palette='coolwarm')
    plt.title(f'Suscripción por "{col}"', fontsize=16)
    plt.xlabel('Número de Clientes', fontsize=12)
    plt.ylabel(col.capitalize(), fontsize=12)
    plt.legend(title='Suscripción')
    plt.show()
```

A este análisis añado una matriz de correlación que muestre la relación lineal entre las variables numéricas para tener una visión general de las posibles interdependencias en los datos. 

```python
# Matriz de Correlación

# Selecciono solo las columnas numéricas para el cálculo
numeric_df = df_completo.select_dtypes(include=np.number)

# Calculo la matriz de correlación
correlation_matrix = numeric_df.corr()

# Creo el mapa de calor (heatmap)
plt.figure(figsize=(18, 15))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt='.2f', linewidths=.5)
plt.title('Matriz de Correlación de Variables Numéricas', fontsize=20)
plt.show()
```
